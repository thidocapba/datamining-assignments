**Bước 1: Khai báo thư viện**

```python
import pandas as pd 
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

**Buớc 2: Đọc dữ liệu**

```python
df = pd.read_csv("bank_data.csv")
```

**Bước 3: Hiểu dữ liệu**

```python
# Xem 5 dòng dữ liệu đầu tiên
df.head()
```

```python
# Xem 5 dòng dữ liệu cuối cùng
df.tail()
```

```python
# Số lượng mẫu và số thuộc tính
print(df.shape)

# Số lượng mẫu
print(df.shape[0])

# Số lượng thuộc tính
print(df.shape[1])
```

```python
# Thống kê các giá trị định lượng
df.describe()
```

```python
# Hiển thị kiểu dữ liệu của các thuộc tính
df.dtypes
```

```python
# Đổi dữ liệu từ dạng định danh (object) về dạng số
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
df['sex'] = lb_make.fit_transform(df['sex'])
df['region'] = lb_make.fit_transform(df['region'])
df['married'] = lb_make.fit_transform(df['married'])
df['car'] = lb_make.fit_transform(df['car'])
df['save_act'] = lb_make.fit_transform(df['save_act'])
df['current_act'] = lb_make.fit_transform(df['current_act'])
df['mortgage'] = lb_make.fit_transform(df['mortgage'])
df['pep'] = lb_make.fit_transform(df['pep'])
df.head(10)
```

**Bước 4: Xây dựng model dự đoán khả năng trả nợ của khách hàng**

```python
# Xác định thuộc tính mô tả X và thuộc tính dự đoán y
features = ['age', 'sex', 'region', 'income', 'married', 'children', 'car', 'save_act', 'current_act', 'mortgage']
target = ['pep']
X = df[features]
y = df[target]
print(X)
print(y)
```

```python
# Chia bộ dữ liệu thành 2 tập train và test theo tỉ lệ 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
# Khai báo mô hình cây quyết định
model = tree.DecisionTreeClassifier(criterion = "entropy",
                                   random_state = 100,
                                   max_depth = 3,
                                   min_samples_leaf = 5)
```

```python
# Thực thi mô hình
model = model.fit(X_train, y_train)
```

**Bước 5: Xác định độ chính xác của mô hình**

```python
# Kiểm thử mô hình
y_pred = model.predict(X_test)
print("Giá trị nhãn mô hình dự đoán được:\n ", y_pred)
```

```python
# Xác định ma trận nhầm lẫn
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
```

```python
# Độ chính xác của mô hình
print("Accuracy : ", accuracy_score(y_test, y_pred) * 100)
```

**Bước 6: Sử dụng mô hình**

```python
# Sử dụng mô hình dự đoán khả năng trả nợ của khách hàng có các thông tin sau
# age: 42, sex: female 1, region: town 3, income: 30537, married: yes 1, children: 2, car: yes 1,
# save_act: yes 1, current_act: no 0, mortgage: yes 1.

x = [[42, 1, 3, 30537, 1, 2, 1, 1, 0, 1]]
y = model.predict(x)
if y == 1:
    print("Khách hàng có khả năng trả nợ")
else:
    print("Khách hàng không có khả năng trả nợ")
```
